```python
import pandas as pd
import ast
import os
```

```python
os.chdir("../Data")
```

```python
df0 = pd.read_csv("comments_text.csv")
print(df0.shape)
df0.head()
```

```python
with open('nlp_response/postman_delayed_4sec.txt', 'r') as f:
    response = f.read().replace('\n', '')
    
response = response[:-1]
response = '[' + response + ']'
nlp_list = ast.literal_eval(response)
```

```python
df1 = pd.DataFrame(nlp_list)
print(df1.shape)
df1.head()
```

```python
# Adding text classification to the comments dataframe
df2 = pd.concat([df1, df0.head(1000)], axis=1)
df2[435:]
```

```python
df_comments = df2.copy()
```

```python
# rounding values to the nearest integer
df_comments[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']] = \
df_comments[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']].round()
df_comments.head()
```

```python
# change the type of the rounded values to boolean
df_comments['defense_AH'] = df_comments['defense_AH'].astype('bool')
df_comments['support_AH'] = df_comments['support_AH'].astype('bool')
df_comments['offense_AH'] = df_comments['offense_AH'].astype('bool')
df_comments['defense_against_AH'] = df_comments['defense_against_AH'].astype('bool')
```

## Save the Classified DataFrame

```python
df_comments.to_csv('instagram_delayed_4sec.csv', index=False)
```
