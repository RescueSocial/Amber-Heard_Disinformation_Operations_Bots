## Add Metrics to the tweets data

```python
zfile = ZipFile('Data_Kaggle.zip', 'r')

filename = 'Data_Kaggle/2018 Tweets_Metrics.csv'
with zfile.open(filename) as f:
    metrics_18 = pd.read_csv(StringIO(f.read().decode()), lineterminator='\n')

filename = 'Data_Kaggle/2019 Tweets_Metrics.csv'
with zfile.open(filename) as f:
    metrics_19 = pd.read_csv(StringIO(f.read().decode()), lineterminator='\n')

filename = 'Data_Kaggle/2020 Tweets_Metrics.csv'
with zfile.open(filename) as f:
    metrics_20 = pd.read_csv(StringIO(f.read().decode()), lineterminator='\n')
    
filename = 'Data_Kaggle/2021 Tweets_Metrics.csv' 
with zfile.open(filename) as f:
    metrics_21 = pd.read_csv(StringIO(f.read().decode()), lineterminator='\n')
    
filename = 'Data_Kaggle/2022 Tweets_Metrics.csv'
with zfile.open(filename) as f:
    metrics_22 = pd.read_csv(StringIO(f.read().decode()), lineterminator='\n')
    
zfile.close()
```

```python
df_metrics = pd.concat([metrics_18, metrics_19, metrics_20, metrics_21, metrics_22])
print(df_metrics.shape)
df_metrics.head(1)
```

```python
df_metrics.drop(columns=['created_at', 'text', 'author.id', 'author.username'], inplace=True)
```

```python
# Rename Columns 
df_metrics.rename(columns={'public_metrics.retweet_count': 'retweet_count',
                          'public_metrics.reply_count': 'reply_count',
                          'public_metrics.like_count': 'like_count',
                          'public_metrics.quote_count': 'quote_count'}, inplace=True);
```

#### Retweets
A count of how many times the Tweet has been Retweeted. <br>
Please note: This does not include Quote Tweets (â€œRetweets with commentâ€). 

#### Quote Tweets
A count of how many times the Tweet has been Retweeted with a new comment (message). <br>
Please note: This does not include Retweets. <br>


**To get the â€œRetweets and commentsâ€ total as displayed on the Twitter clients, simply add retweet_count and quote_count. <br>
This will return the total count of Quote Tweets. There are no Quote Tweets from a paid context so all Quote Tweets are organic.**

```python
df_metrics['total_retweets'] = df_metrics['retweet_count'] + df_metrics['quote_count']
```

## Wrangle Users Data

```python
users['user.id_str'].value_counts().head()  # [users['user.id_str'].value_counts() < 5]
```

> **NOTE:**
> - Duplicates in the user id, is caused by users who changed their info
> - 'username', 'user_location', 'profile_image_url', 'profile_banner_url' can be changed
> - or changes in 'n_followers', 'n_friends' or 'n_statuses' with time

### Drop Duplicates Where Rows has the most nans

https://stackoverflow.com/questions/43769693/drop-duplicates-in-a-dataframe-keeping-the-row-with-the-least-nulls

```python
users_cleaned = users.assign(counts=users.count(axis=1)).sort_values(['user.id_str', 'counts'])\
.drop_duplicates('user.id_str', keep='last').drop('counts', axis=1)

users_cleaned['user.created_at'] = pd.to_datetime(users_cleaned["user.created_at"], format="%Y-%m-%d %H:%M:%S+00:00") 
users_cleaned.sort_values('user.id_str', inplace=True)
# users_cleaned['user.id_str'] = users_cleaned['user.id_str'].astype('str')
print(users_cleaned.shape)
users_cleaned.head(2)
```

> **NOTE:**
> - This is the same Count we got, when using the top user metrics approach
> - The only difference is, that we did not keep the data that can be changed in the 3rd approach.
> - 'username', 'user_location', 'profile_image_url', 'profile_banner_url' can be changed

You can sort the dataFrame by count and then remove duplicates. I think it's easier:

```python
# tweets.columns
```

```python
user_cols = ['user_id', 'user_created_at', 'username', 'user_screen_name', 'n_followers', 'n_friends', 
             'n_statuses', 'user_url', 'user_description', 'user_location', 
             'profile_banner_url', 'profile_image_url', 'is_verified']
```

```python
users2 = tweets.sort_values(['n_followers', 'n_friends', 'n_statuses'], 
                            ascending=[False, False, False]).drop_duplicates(['user_id'])

users3 = users2[user_cols] 
users3['user_id'] = users3['user_id'].astype('int')
users3.sort_values('user_id', inplace=True)
print(users3.shape)
users3.head()
```

```python
users4 = tweets.groupby(['user_id', 'user_created_at'])\
.agg({'n_followers': 'max', 'n_friends': 'max', 'n_statuses': 'max'}).reset_index()
users4.sort_values(['n_followers', 'n_friends', 'n_statuses'], ascending=[False, False, False], inplace=True)
users4.sort_values('user_created_at', inplace=True)
print(users4.shape)
users4.head()
```
