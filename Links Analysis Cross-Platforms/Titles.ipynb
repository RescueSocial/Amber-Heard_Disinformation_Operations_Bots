```python
def deEmojify(text):
    regrex_pattern = re.compile(
        pattern="["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "]+",
        flags=re.UNICODE,
    )
    return regrex_pattern.sub(r"", text)
```

```python
"""def get_title(url):
    try:
        reqs = requests.get(url)
        soup = BeautifulSoup(reqs.text, "html.parser")

        return soup.find("title").get_text()
    except:
        return "error"""


def get_title(url):
    url = deEmojify(url)
    if "https://" not in url:
        url = "https://" + url
    br = Browser()
    br.set_handle_equiv(False)
    br._factory.is_html = True

    br.addheaders = [
        (
            "User-agent",
            "Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.9.0.1) Gecko/2008071615 Fedora/3.0.1-1.fc9 Firefox/3.0.1",
        )
    ]

    br.set_handle_robots(False)

    try:
        br.open(url)
        return br.title()
    except HTTPError:
        return "error http"

    except:
        return "error"
```

```python
def saving(platform, df):
    name = df.name
    df.to_csv(platform + name + "_title" +".csv", index=False)
    print(platform + name + "_title"+ ".csv  " + "SAVED!")
```

```python
def df_with_title(df, platform):
    df["title"] = df["link"].apply(lambda x: get_title(x))
    saving(platform, df)
```
